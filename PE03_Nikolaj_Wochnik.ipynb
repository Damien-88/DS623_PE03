### Import Libraries

```python
import numpy as np
import math as m
```

### L1 Norm Function

```python
def l1_norm(v):
    """Calculate the Manhattan norm of a vector."""
    total = 0 # Initialize the total variable.

    # Loop through each element in the vector and add its absolute value to total.
    for e in v:
        total += abs(e)

    return total
```

### L2 Norm Function

```python
def l2_norm(v):
    """Calculate the Euclidean norm of a vector."""
    total = 0 # Initialize the total variable

    # Loop through each element in the vector and add its square to total.
    for e in v:
        total += e**2

    return m.sqrt(total) # Return square root of total of squared elements.
```

### L1 Distance Funciton

```python
def l1_distance(v1, v2):
    """Calculate the Manhattan distance between two vectors."""
    distance = 0 # Intitalize the distance variable

    # Loop through each pair of elements in the two vectors.
    # Add the absolute difference of each to the distance vaiable.
    for e1, e2 in zip(v1, v2):
        distance += abs(e1 - e2)

    return distance
```

### L2 Distance Function

```python
def l2_distance(v1, v2):
    """Calculate the Euclidean distance between two vectors."""
    distance = 0 # Initialize the distance variable

    # Loop through each pair of elements in the two vectors.
    # Add the square of the difference of each to the distance variable.
    for e1, e2 in zip(v1, v2):
        distance += (e1 - e2)**2

    return m.sqrt(distance) # Return square root of total of squared differences.
```

### Dot Product Function

```python
def dot_product(v1, v2):
    """Calculate the dot product of two vectors."""
    total = 0 # Initialize the total variable

    # Loop though each pair of elements in the two vectors.
    # Add the product of each to the total variable.
    for e1, e2 in zip(v1, v2):
        total += e1 * e2

    return total # Return total of products of each element pair from two vectors.
```

### Area of Parallelogram Function

```python
def parallelogram_area(v1, v2):
    """Calculate the area of a parallelogram formed by two vectors."""
    dot = dot_product(v1, v2) # Calculate dot product of two vectors
    norm1 = l2_norm(v1) # Calculate Euclidean norm of first vector
    norm2 = l2_norm(v2) # Calculate Euclidean norm of second vector

    # Calculate angle between two vectors using dot product and norms
    angle = m.acos(dot /norm1 / norm2) 

    # Calculate area of parallelogram
    area = norm1 * norm2 * m.sin(angle) 

    return area # Return area of parallelogram formed by two vectors.
```

### Enter Vectors

```python
v1 = np.array([2, 1, -3])
v2 = np.array([-3, 0, 5])
```

### L1 Caclulations

```python
l1_v1 = l1_norm(v1)             # Calculate Manhattan norm of first vector
l1_v2 = l1_norm(v2)             # Calculate Manhattan norm of second vector
l1_dist = l1_distance(v1, v2)   # Calculate Manhattan distance between two vectors
l1_sum = l1_norm(v1 + v2)       # Calculate Manhattan norm of the sum of two vectors

print(f"a. L1 norm of v1: \n\t{l1_v1}")
print(f"b. L1 norm of v2: \n\t{l1_v2}")
print(f"c. L1 distance between v1 and v2: \n\t{l1_dist}")
print("d. Triangular inequality: ")
print(f"\t{l1_v1} + {l1_v2} >= {l1_sum}; {l1_v1 + l1_v2 >= l1_sum}")
```

### L2 Calculations

```python
l2_v1 = l2_norm(v1)             # Calculate Euclidean norm of first vector
l2_v2 = l2_norm(v2)             # Calculate Euclidean norm of second vector
l2_dist = l2_distance(v1, v2)   # Calculate Euclidean distance between two vectors               
l2_sum = l2_norm(v1 + v2)       # Calculate Euclidean norm of the sum of two vectors

print(f"e. L2 norm of v1: \n\t{l2_v1:.3f}")
print(f"f. L2 norm of v2: \n\t{l2_v2:.3f}")
print(f"g. L2 distance between v1 and v2: \n\t{l2_dist:.3f}")
print(f"h. Triangular inequality: ")
print(f"\t{l2_v1:.3f} + {l2_v2:.3f} >= {l2_sum:.3f}; {l2_v1 + l2_v2 >= l2_sum}")
```

### Area of the Parallelogram

```python
area = parallelogram_area(v1, v2) # Calculate area of parallelogram formed by two vectors

print(f"i. Area of parallelogram formed by v1 and v2: \n\t{area:.3f}")
```
